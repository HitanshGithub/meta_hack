## 5) Start OpenEnv server inside notebook

```python
import subprocess
import time

# Important: avoid stdout=PIPE here, otherwise uvicorn logs can fill the pipe and block requests.
server = subprocess.Popen(
    ["python", "-m", "uvicorn", "server.app:app", "--host", "0.0.0.0", "--port", "7860"],
    stdout=subprocess.DEVNULL,
    stderr=subprocess.STDOUT,
)
time.sleep(6)
print("Server PID:", server.pid)
!curl -s http://127.0.0.1:7860/health
```

## 6) Run GRPO training (Dr. GRPO + KL penalty)
Uses Dr. GRPO loss with KL penalty to ensure non-zero loss and stable gradients.

```python
!mkdir -p artifacts/grpo_kaggle_run
!python train_grpo.py \
  --env-base-url http://127.0.0.1:7860 \
  --model-name Qwen/Qwen2.5-0.5B-Instruct \
  --num-samples 120 \
  --num-train-epochs 2 \
  --per-device-train-batch-size 1 \
  --gradient-accumulation-steps 4 \
  --num-generations 4 \
  --learning-rate 2e-5 \
  --episodes-per-task 1 \
  --max-episode-steps 3 \
  --eval-tasks-per-difficulty 1 \
  --env-timeout-seconds 120 \
  --env-max-retries 3 \
  --loss-type dr_grpo \
  --beta 0.04 \
  --max-completion-length 256 \
  --max-new-tokens 256 \
  --output-dir artifacts/grpo_kaggle_run 2>&1 | tee artifacts/grpo_kaggle_run/training.log
```

## 7) View training outputs

```python
from IPython.display import Image, Markdown, display
import json, os

plots_dir = 'artifacts/grpo_kaggle_run/plots'
logs_dir = 'artifacts/grpo_kaggle_run/logs'

for title, fname in [
    ('Reward Curve', 'reward_curve.png'),
    ('Training Loss Curve', 'loss_curve.png'),
    ('Auxiliary Loss Curve', 'aux_loss_curve.png'),
    ('Trainer Aux Loss Curve', 'trainer_aux_loss_curve.png'),
]:
    path = os.path.join(plots_dir, fname)
    if os.path.exists(path):
        display(Markdown(f'### {title}'))
        display(Image(filename=path))
    else:
        print(f'[SKIP] {title} not found at {path}')

ba_path = os.path.join(logs_dir, 'before_after.md')
if os.path.exists(ba_path):
    display(Markdown('### Performance Evaluation'))
    display(Markdown(open(ba_path, 'r', encoding='utf-8').read()))

summary_path = os.path.join(logs_dir, 'training_summary.json')
if os.path.exists(summary_path):
    summary = json.load(open(summary_path, 'r', encoding='utf-8'))
    display(summary)
```
